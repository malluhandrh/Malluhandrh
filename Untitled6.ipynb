```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio
import ipywidgets as w
from IPython.display import display, clear_output
from google.colab import output
output.enable_custom_widget_manager()

# usar o renderer correto no Colab
pio.renderers.default = "colab"

# ===== formatadores de moeda =====
def fmt_brl(val):
    try:
        return f"R$ {float(val):,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
    except:
        return "R$ 0,00"
def fmt_list(vals):
    return [fmt_brl(v) for v in vals]

# ===== widgets =====
unidades = []
if "UNIDADE" in df.columns:
    unidades = sorted([u for u in df["UNIDADE"].dropna().astype(str).str.strip().unique().tolist() if u != ""])

meses_ord = df.sort_values("Mes_dt")["Mes_fmt"].unique().tolist()

w_unid = w.SelectMultiple(
    options=unidades,
    value=tuple([]),  # vazio = todas
    description='Unidade(s):',
    layout=w.Layout(width='360px', height='160px')
)
w_ini = w.Dropdown(options=meses_ord, value=meses_ord[0], description='InÃ­cio:')
w_fim = w.Dropdown(options=meses_ord, value=meses_ord[-1], description='Fim:')

filtros = w.HBox([w_unid, w.VBox([w_ini, w_fim])])
display(filtros)

# >>> use DOIS outputs: um para cards, outro para grÃ¡ficos
out_cards = w.Output()
out_fig   = w.Output()
display(out_cards)
display(out_fig)

def atualizar(*args):
    # limpa e desenha os CARDS sempre no out_cards
    with out_cards:
        clear_output(wait=True)

        df_f = df.copy()

        # perÃ­odo (corrige inversÃ£o)
        meses_sorted = df.sort_values("Mes_dt")["Mes_fmt"].unique().tolist()
        i0 = meses_sorted.index(w_ini.value)
        i1 = meses_sorted.index(w_fim.value)
        start = meses_sorted[min(i0, i1)]
        end   = meses_sorted[max(i0, i1)]

        m0 = pd.to_datetime("01/" + start, dayfirst=True, errors="coerce")
        m1 = pd.to_datetime("01/" + end,   dayfirst=True, errors="coerce")
        df_f = df_f[(df_f["Mes_dt"] >= m0) & (df_f["Mes_dt"] <= m1)]

        # unidades (multi)
        sel_unids = list(w_unid.value)
        if sel_unids and "UNIDADE" in df_f.columns:
            df_f["UNIDADE"] = df_f["UNIDADE"].astype(str).str.strip()
            sel_unids = [u.strip() for u in sel_unids]
            df_f = df_f[df_f["UNIDADE"].isin(sel_unids)]

        if df_f.empty:
            print("ðŸ”Ž Nenhum dado para o filtro escolhido.")
            return

        # custo total
        df_f["CUSTO_TOTAL"] = 0.0
        for c in ["CUSTO DO SERV. PRESTADO","CUSTO FIXOS","CUSTOS VARIAVEIS"]:
            if c in df_f.columns:
                df_f["CUSTO_TOTAL"] = df_f["CUSTO_TOTAL"].fillna(0) + df_f[c].fillna(0)

        # agregados por mÃªs
        cols_agg = {}
        if "RECEITA BRUTA" in df_f.columns: cols_agg["RECEITA BRUTA"] = "sum"
        cols_agg["CUSTO_TOTAL"] = "sum"
        if "LUCRO BRUTO" in df_f.columns: cols_agg["LUCRO BRUTO"] = "sum"

        grp = (df_f.groupby("Mes_fmt", as_index=False)
                     .agg(cols_agg)
                     .rename(columns={"RECEITA BRUTA":"Receita",
                                      "CUSTO_TOTAL":"Custos",
                                      "LUCRO BRUTO":"Lucro"})
                     .sort_values("Mes_fmt"))

        # receita por unidade
        if "UNIDADE" in df_f.columns and "RECEITA BRUTA" in df_f.columns:
            base_uni = (df_f.groupby("UNIDADE", as_index=False)["RECEITA BRUTA"]
                          .sum().rename(columns={"RECEITA BRUTA":"Receita"})
                          .sort_values("Receita", ascending=False))
        else:
            base_uni = pd.DataFrame(columns=["UNIDADE","Receita"])

        # KPIs
        receita_total = float(grp["Receita"].sum()) if "Receita" in grp else 0.0
        custos_total  = float(grp["Custos"].sum())  if "Custos"  in grp else 0.0
        lucro_total   = float(grp["Lucro"].sum())   if "Lucro"   in grp else 0.0
        receita_media_unidade = float(base_uni["Receita"].mean()) if not base_uni.empty else 0.0

        # ===== cards 2x2 =====
        cards = make_subplots(
            rows=2, cols=2,
            specs=[[{"type":"indicator"},{"type":"indicator"}],
                   [{"type":"indicator"},{"type":"indicator"}]],
            subplot_titles=("Receita Total","Custos Totais","Lucro Total","Receita MÃ©dia/Unidade")
        )
        cards.add_trace(go.Indicator(
            mode="number",
            value=receita_total,
            number=dict(valueformat=",.2f", font=dict(size=32, color="green")),
            title={"text": fmt_brl(receita_total)}), 1,1)

        cards.add_trace(go.Indicator(
            mode="number",
            value=custos_total,
            number=dict(valueformat=",.2f", font=dict(size=32, color="red")),
            title={"text": fmt_brl(custos_total)}), 1,2)

        cards.add_trace(go.Indicator(
            mode="number",
            value=lucro_total,
            number=dict(valueformat=",.2f", font=dict(size=32, color="blue")),
            title={"text": fmt_brl(lucro_total)}), 2,1)

        cards.add_trace(go.Indicator(
            mode="number",
            value=receita_media_unidade,
            number=dict(valueformat=",.2f", font=dict(size=32, color="orange")),
            title={"text": fmt_brl(receita_media_unidade)}), 2,2)

        cards.update_layout(
            height=400, width=980,
            paper_bgcolor="#F7F9FB",
            font=dict(family="Arial", size=14),
            margin=dict(t=40, b=20)
        )

        display(cards)

    # limpa e desenha os GRÃFICOS sempre no out_fig
    with out_fig:
        clear_output(wait=True)

        # reusa df_f, grp, base_uni calculados acima? recalcula rÃ¡pido:
        df_f2 = df.copy()
        i0 = meses_sorted.index(start); i1 = meses_sorted.index(end)
        m0 = pd.to_datetime("01/" + meses_sorted[min(i0, i1)], dayfirst=True, errors="coerce")
        m1 = pd.to_datetime("01/" + meses_sorted[max(i0, i1)], dayfirst=True, errors="coerce")
        df_f2 = df_f2[(df_f2["Mes_dt"] >= m0) & (df_f2["Mes_dt"] <= m1)]
        if sel_unids and "UNIDADE" in df_f2.columns:
            df_f2["UNIDADE"] = df_f2["UNIDADE"].astype(str).str.strip()
            df_f2 = df_f2[df_f2["UNIDADE"].isin(sel_unids)]

        df_f2["CUSTO_TOTAL"] = 0.0
        for c in ["CUSTO DO SERV. PRESTADO","CUSTO FIXOS","CUSTOS VARIAVEIS"]:
            if c in df_f2.columns:
                df_f2["CUSTO_TOTAL"] = df_f2["CUSTO_TOTAL"].fillna(0) + df_f2[c].fillna(0)

        cols_agg2 = {}
        if "RECEITA BRUTA" in df_f2.columns: cols_agg2["RECEITA BRUTA"] = "sum"
        cols_agg2["CUSTO_TOTAL"] = "sum"
        if "LUCRO BRUTO" in df_f2.columns: cols_agg2["LUCRO BRUTO"] = "sum"

        grp2 = (df_f2.groupby("Mes_fmt", as_index=False)
                       .agg(cols_agg2)
                       .rename(columns={"RECEITA BRUTA":"Receita",
                                        "CUSTO_TOTAL":"Custos",
                                        "LUCRO BRUTO":"Lucro"})
                       .sort_values("Mes_fmt"))
        if "UNIDADE" in df_f2.columns and "RECEITA BRUTA" in df_f2.columns:
            base_uni2 = (df_f2.groupby("UNIDADE", as_index=False)["RECEITA BRUTA"]
                           .sum().rename(columns={"RECEITA BRUTA":"Receita"})
                           .sort_values("Receita", ascending=False))
        else:
            base_uni2 = pd.DataFrame(columns=["UNIDADE","Receita"])

        fig = make_subplots(
            rows=2, cols=2,
            subplot_titles=("Receita Bruta", "Custos", "Lucro Bruto", "Receita por Unidade"),
            specs=[[{"type":"scatter"}, {"type":"bar"}],
                   [{"type":"scatter"}, {"type":"bar"}]]
        )

        if "Receita" in grp2.columns and not grp2.empty:
            fig.add_trace(
                go.Scatter(x=grp2["Mes_fmt"], y=grp2["Receita"],
                           mode="lines+markers", name="Receita",
                           text=fmt_list(grp2["Receita"]), textposition="top center"),
                row=1, col=1
            )

        if "Custos" in grp2.columns and not grp2.empty:
            fig.add_trace(
                go.Bar(x=grp2["Mes_fmt"], y=grp2["Custos"], name="Custos",
                       text=fmt_list(grp2["Custos"]), textposition="outside"),
                row=1, col=2
            )

        if "Lucro" in grp2.columns and not grp2.empty:
            fig.add_trace(
                go.Scatter(x=grp2["Mes_fmt"], y=grp2["Lucro"],
                           mode="lines+markers", name="Lucro",
                           text=fmt_list(grp2["Lucro"]), textposition="top center"),
                row=2, col=1
            )

        if not base_uni2.empty:
            fig.add_trace(
                go.Bar(x=base_uni2["UNIDADE"], y=base_uni2["Receita"], name="Receita por Unidade",
                       text=fmt_list(base_uni2["Receita"]), textposition="outside"),
                row=2, col=2
            )

        fig.update_layout(
            height=720, width=980, showlegend=False,
            title_text="ðŸ“Š Dashboard DRE (interativo com filtros)",
            title_x=0.5,
            paper_bgcolor="#F7F9FB", plot_bgcolor="white",
            font=dict(family="Arial", size=14),
            margin=dict(t=60, b=40)
        )
        display(fig)

# conecta eventos e renderiza
for wid in (w_unid, w_ini, w_fim):
    wid.observe(atualizar, names='value')
atualizar()
```

```python
import numpy as np, re, unicodedata

ANO_REFERENCIA = 2025

def remove_acentos(s: str) -> str:
    return ''.join(ch for ch in unicodedata.normalize('NFKD', s) if not unicodedata.combining(ch))

def normaliza_txt(s: str) -> str:
    s = str(s).strip()
    s = remove_acentos(s)
    return re.sub(r"\s+", " ", s)

map_pt = {
    "JANEIRO":1,"JAN":1, "FEVEREIRO":2,"FEV":2, "MARCO":3,"MAR":3,
    "ABRIL":4,"ABR":4, "MAIO":5,"MAI":5, "JUNHO":6,"JUN":6,
    "JULHO":7,"JUL":7, "AGOSTO":8,"AGO":8, "SETEMBRO":9,"SET":9,
    "OUTUBRO":10,"OUT":10, "NOVEMBRO":11,"NOV":11, "DEZEMBRO":12,"DEZ":12,
}

def parse_mes(raw):
    s_raw = str(raw)
    s = normaliza_txt(s_raw).upper()
    if s in {"", "NAN", "MES"}:
        return pd.NaT
    if s in map_pt:
        return pd.Timestamp(year=ANO_REFERENCIA, month=map_pt[s], day=1)
    m = re.fullmatch(r"(\d{1,2})\s*/\s*(\d{2,4})", s)
    if m:
        mm = int(m.group(1)); aa = int(m.group(2))
        if aa < 100: aa = 2000 + aa
        try: return pd.Timestamp(year=aa, month=mm, day=1)
        except: return pd.NaT
    dt = pd.to_datetime(s_raw, dayfirst=True, errors="coerce")
    if pd.notna(dt):
        return pd.Timestamp(year=dt.year, month=dt.month, day=1)
    return pd.NaT

def to_float_br(x):
    if pd.isna(x): return np.nan
    if isinstance(x, (int, float, np.number)): return float(x)
    s = str(x)
    s = s.replace("R$", "").replace(" ", "").replace("â€“", "-").replace("â€”", "-")
    s = re.sub(r"[^0-9,\.\-]", "", s)
    if "," in s: s = s.replace(".", "").replace(",", ".")
    return pd.to_numeric(s, errors="coerce")

df = df.copy()
df.columns = [normaliza_txt(c) for c in df.columns]

ren = {
    "Mes":"Mes","UNIDADE":"UNIDADE","RECEITA BRUTA":"RECEITA BRUTA",
    "CUSTO DO SERV. PRESTADO":"CUSTO DO SERV. PRESTADO","CUSTO FIXOS":"CUSTO FIXOS",
    "CUSTOS VARIAVEIS":"CUSTOS VARIAVEIS","Lucro Bruto":"LUCRO BRUTO"
}
df = df.rename(columns={k:v for k,v in ren.items() if k in df.columns})

df["Mes_dt"] = pd.to_datetime(df["Mes"].apply(parse_mes), errors="coerce")
df = df[df["Mes_dt"].notna()].copy()
df["Mes_fmt"] = df["Mes_dt"].dt.strftime("%m/%y")

num_cols = [c for c in ["RECEITA BRUTA","CUSTO DO SERV. PRESTADO","CUSTO FIXOS",
                        "CUSTOS VARIAVEIS","LUCRO BRUTO"] if c in df.columns]
for c in num_cols:
    df[c] = df[c].apply(to_float_br)
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import ipywidgets as w
from IPython.display import display, clear_output
import plotly.io as pio
pio.renderers.default = "colab"

def fmt_brl(val):
    try: return f"R$ {float(val):,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
    except: return "R$ 0,00"
def fmt_list(vals): return [fmt_brl(v) for v in vals]

# widgets
unidades = sorted(df["UNIDADE"].dropna().astype(str).str.strip().unique().tolist()) if "UNIDADE" in df.columns else []
meses_ord = df.sort_values("Mes_dt")["Mes_fmt"].unique().tolist()

w_unid = w.SelectMultiple(options=unidades, value=tuple([]), description='Unidade(s):')
w_ini = w.Dropdown(options=meses_ord, value=meses_ord[0], description='InÃ­cio:')
w_fim = w.Dropdown(options=meses_ord, value=meses_ord[-1], description='Fim:')
display(w.HBox([w_unid, w.VBox([w_ini, w_fim])]))

out_cards, out_fig = w.Output(), w.Output()
display(out_cards); display(out_fig)

# guardar Ãºltimas figs p/ exportar
last_cards, last_fig = None, None

def atualizar(*args):
    global last_cards, last_fig
    with out_cards: clear_output(wait=True)
    with out_fig: clear_output(wait=True)

    df_f = df.copy()
    i0, i1 = meses_ord.index(w_ini.value), meses_ord.index(w_fim.value)
    start, end = meses_ord[min(i0,i1)], meses_ord[max(i0,i1)]
    m0, m1 = pd.to_datetime("01/"+start, dayfirst=True), pd.to_datetime("01/"+end, dayfirst=True)
    df_f = df_f[(df_f["Mes_dt"]>=m0)&(df_f["Mes_dt"]<=m1)]
    if w_unid.value: df_f = df_f[df_f["UNIDADE"].astype(str).str.strip().isin(list(w_unid.value))]
    if df_f.empty:
        print("Sem dados para este filtro."); return

    df_f["CUSTO_TOTAL"]=0
    for c in ["CUSTO DO SERV. PRESTADO","CUSTO FIXOS","CUSTOS VARIAVEIS"]:
        if c in df_f.columns: df_f["CUSTO_TOTAL"]+=df_f[c].fillna(0)
    grp = df_f.groupby("Mes_fmt",as_index=False).agg({
        "RECEITA BRUTA":"sum","CUSTO_TOTAL":"sum","LUCRO BRUTO":"sum"
    }).rename(columns={"RECEITA BRUTA":"Receita","CUSTO_TOTAL":"Custos","LUCRO BRUTO":"Lucro"})
    base_uni = df_f.groupby("UNIDADE",as_index=False)["RECEITA BRUTA"].sum().rename(columns={"RECEITA BRUTA":"Receita"}) if "UNIDADE" in df_f.columns else pd.DataFrame()

    # KPIs
    receita_total, custos_total, lucro_total = grp["Receita"].sum(), grp["Custos"].sum(), grp["Lucro"].sum()
    receita_media_unidade = base_uni["Receita"].mean() if not base_uni.empty else 0

    cards = make_subplots(rows=2, cols=2, specs=[[{"type":"indicator"}]*2]*2,
        subplot_titles=("Receita Total","Custos Totais","Lucro Total","Receita MÃ©dia/Unidade"))
    cards.add_trace(go.Indicator(mode="number", value=receita_total,
        number=dict(valueformat=",.2f",font=dict(size=32,color="green")),
        title={"text":fmt_brl(receita_total)}),1,1)
    cards.add_trace(go.Indicator(mode="number", value=custos_total,
        number=dict(valueformat=",.2f",font=dict(size=32,color="red")),
        title={"text":fmt_brl(custos_total)}),1,2)
    cards.add_trace(go.Indicator(mode="number", value=lucro_total,
        number=dict(valueformat=",.2f",font=dict(size=32,color="blue")),
        title={"text":fmt_brl(lucro_total)}),2,1)
    cards.add_trace(go.Indicator(mode="number", value=receita_media_unidade,
        number=dict(valueformat=",.2f",font=dict(size=32,color="orange")),
        title={"text":fmt_brl(receita_media_unidade)}),2,2)
    cards.update_layout(height=400,width=980,paper_bgcolor="#F7F9FB")
    display(cards)

    fig = make_subplots(rows=2, cols=2,
        subplot_titles=("Receita Bruta","Custos","Lucro Bruto","Receita por Unidade"),
        specs=[[{"type":"scatter"},{"type":"bar"}],[{"type":"scatter"},{"type":"bar"}]])
    fig.add_trace(go.Scatter(x=grp["Mes_fmt"],y=grp["Receita"],mode="lines+markers",
        text=fmt_list(grp["Receita"]),textposition="top center"),1,1)
    fig.add_trace(go.Bar(x=grp["Mes_fmt"],y=grp["Custos"],
        text=fmt_list(grp["Custos"]),textposition="outside"),1,2)
    fig.add_trace(go.Scatter(x=grp["Mes_fmt"],y=grp["Lucro"],mode="lines+markers",
        text=fmt_list(grp["Lucro"]),textposition="top center"),2,1)
    if not base_uni.empty:
        fig.add_trace(go.Bar(x=base_uni["UNIDADE"],y=base_uni["Receita"],
            text=fmt_list(base_uni["Receita"]),textposition="outside"),2,2)
    fig.update_layout(height=720,width=980,showlegend=False,
        paper_bgcolor="#F7F9FB",plot_bgcolor="white",
        title_text="ðŸ“Š Dashboard DRE",title_x=0.5)
    display(fig)

    last_cards, last_fig = cards, fig

for wid in (w_unid,w_ini,w_fim): wid.observe(atualizar,names='value')
atualizar()
```
